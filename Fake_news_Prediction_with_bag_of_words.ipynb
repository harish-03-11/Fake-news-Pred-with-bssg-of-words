# Converting Word to vector with Bag of n grams/bag of words

#Using Multinomial Naive bayes with Unigram

```python
from sklearn.feature_extraction.text import CountVectorizer
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train,X_test,y_train,y_test = train_test_split(df_new.text,df_new.label,test_size = 0.2, stratify = df_new.label ,random_state=22)
```

```python
from sklearn.naive_bayes import MultinomialNB
from sklearn.metrics import classification_report
from sklearn.pipeline import Pipeline
```

```python
model = Pipeline([
    ('vectorizer', CountVectorizer()),
    ('Multi NB', MultinomialNB())

])
```

```python
model.fit(X_train,y_train)
```

```python
y_pred = model.predict(X_train)
```

```python
print(classification_report(y_pred,y_train))
```

#Using Multinomial Naive bayes with Unigram,Bigram and Trigram

```python
model_3gram = Pipeline([
    ('vectorizer', CountVectorizer(ngram_range = (1,3))),
    ('Multi NB', MultinomialNB())

])
```

```python
model_3gram.fit(X_train,y_train)
```

```python
y3_pred = model_3gram.predict(X_train)
```

```python
print(classification_report(y3_pred,y_train))
```

#Using K Nearest neighbors Classifier

```python
from sklearn.neighbors import KNeighborsClassifier
```

```python
model_knn = Pipeline([
    ('vectorizer', CountVectorizer(ngram_range = (1,3))),
    ('KNN', KNeighborsClassifier(n_neighbors=10,metric='euclidean'))

])
```

```python
model_knn.fit(X_train,y_train)
```

```python
y_knn_pred = model_knn.predict(X_train)
```

```python
print("report For KNN")
print(" ")
print(classification_report(y_knn_pred,y_train))
```

#Starting with text preprocessing ,removing stop words and training the data


```python
df_new.head(5)
```

```python
import spacy
nlp = spacy.load("en_core_web_sm")
```

```python
def preprocess(text):
  doc = nlp(text)
  filtered = []
  for token in doc:
    if not token.is_stop or token.is_punct:
      filtered.append(token.lemma_)
  return " ".join(filtered)
```

```python
df_new['processed'] = df_new['text'].apply(preprocess)
df_new.head(4)
```

```python
print(df_new['processed'][0])
```

```python
df_pro = df_new
```

```python
df_new.sample(5)
```

#Model training with Random forest Classifier with only Trigram

```python
from sklearn.ensemble import RandomForestClassifier
```

```python
X = df_pro.processed
y = df_pro.label
```

```python
X_train,X_test,y_train,y_test = train_test_split(X,y,test_size = 0.2,random_state=2022,stratify=y)
```

```python
model_rf = Pipeline([
    ('vector', CountVectorizer(ngram_range=(3,3))),
    ('RF', RandomForestClassifier())



])
```

```python
model_rf.fit(X_train,y_train)
```

```python
y_rf_pred = model_rf.predict(X_train)
```

```python
print(classification_report(y_rf_pred,y_train))
```
